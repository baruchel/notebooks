On utilise ici des données mesurées dans le laboratoire de Sciences industrielles du Lycée naval. Les données mesurent (en fonction du temps) l'angle d'une porte équipée d'un système de ferme-porte (_groom_).

### Lecture du fichier de données

On commence par lire les données (enregistrées au format CSV au moment de la mesure) :

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

f = open("/home/thomas/Documents/Thomas/Classes/informatique/td12b/fichier_groom.csv", mode="r")
f.readline() # on "saute" la première ligne
data_x, data_y = [], []
for l in f:
  # la variable l contient la ligne en train d'être lue
  l2 = l.rstrip("\n")      # on supprime le caractère de fin de ligne
  l3 = l2.replace(",",".") # on remplace les virgules par des points
  l4 = l3.split(";")       # on sépare les différents champs
  data_x.append(float(l4[0]))
  data_y.append(float(l4[1]))
f.close()

sz = len(data_x)
```

### Affichage des données brutes

On affiche les données (ainsi qu'un détail pour mieux visualiser le « bruit » :

```python
plt.subplot(121)
plt.plot(data_x, data_y, marker='.', linestyle='', markersize=0.25)
plt.subplot(122)
plt.plot(data_x[55*sz//80:3*sz//4], data_y[55*sz//80:3*sz//4], marker='.', linestyle='', markersize=2)
plt.gcf().set_size_inches(12,6)
```

```python
wsize = 65            # taille fenêtre interpolation
s = np.r_[data_y[wsize-1:0:-1],data_y,data_y[-1:-wsize:-1]]
w = np.hanning(wsize) # préparation de la fenêtre utilisée lors du lissage
data_y_smooth = np.convolve(w/w.sum(),s,mode='valid')
data_y_smooth = data_y_smooth[wsize//2:len(data_y)+wsize//2]

plt.subplot(121)
plt.plot(data_x, data_y, marker='.', linestyle='', markersize=0.25)
plt.plot(data_x, data_y_smooth)
plt.subplot(122)
plt.plot(data_x[13*sz//20:3*sz//4], data_y[13*sz//20:3*sz//4], marker='.', linestyle='', markersize=0.25)
plt.plot(data_x[13*sz//20:3*sz//4], data_y_smooth[13*sz//20:3*sz//4])
plt.gcf().set_size_inches(12,6)
```

### Calcul de la dérivée à partir de données lissées

Le module `numpy` permet une dérivation numérique assez intéressante grâce à la fonction `numpy.gradient`. Cette fonction présente notamment l'avantage de retourner autant de termes, centrés, qu'il y en avait au départ (contrairement à la méthode des différences finies qui en élimine un) ; cela est particulièrement utile pour conserver une cohérence avec la liste des abscisses (et, par suite, pour tracer la dérivée).

```python
# lissage des données
wsize = 65        # taille fenêtre interpolation
wtype = 'hanning' # type fenêtre interpolation
s = np.r_[data_y[wsize-1:0:-1],data_y,data_y[-1:-wsize:-1]]
w = eval('np.'+wtype+'(wsize)')
data_y_smooth = np.convolve(w/w.sum(),s,mode='valid')
data_y_smooth = data_y_smooth[wsize//2:len(data_y)+wsize//2]
# calcul de la dérivée
dx = data_x[1]-data_x[0]
dydx = np.gradient(data_y_smooth, dx)
plt.plot(data_x, data_y, marker='.', linestyle='', markersize=0.25)
plt.plot(data_x, data_y_smooth)
plt.plot(data_x,dydx)
plt.gcf().set_size_inches(12,6)
```
